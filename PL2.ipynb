```python
# CH1903+ coordinates (example values)
zurich_coords = [2682216.85, 1247945.20]
geneva_coords = [2499958.90, 1117839.73]
lugano_coords = [2720030.83, 1098728.49]

# Access individual coordinate values using indexing
print("Zurich x:", zurich_coords[0])
print("Zurich y:", zurich_coords[1])

# A list is suitable for coordinates which come in sets of two, as they allow assigning multiple values to a variable.
```

```python
# Dictionary of cities
cities = {"Zurich": zurich_coords, "Geneva": geneva_coords, "Lugano": lugano_coords}

#print(cities)

# Show what is stored in our keys and values
#print(cities.keys())
#print(cities.values())

# Retrieve coordinates from the dictionary
cz = cities["Zurich"]
cg = cities["Geneva"]
cl = cities["Lugano"]

print(cities["Geneva"])

# Dictionaries allow for easy looking up of different variables without learning all the individual names. 
```

```python
# Dictionary for city distances
dx = {"zl": cz[0] - cl[0], "zg": cz[0] - cg[0], "gl": cg[0] - cl[0]}
dy = {"zl": cz[1] - cl[1], "zg": cz[1] - cg[1], "gl": cg[1] - cl[1]}

# Manhattan distance
d_zl = [(dx["zl"]**2)**0.5, (dy["zl"]**2)**0.5]
d_zg = [(dx["zg"]**2)**0.5, (dy["zg"]**2)**0.5]
d_gl = [(dx["gl"]**2)**0.5, (dy["gl"]**2)**0.5]
manhattan = {"zl": d_zl[0] + d_zl[1], "zg": d_zg[0] + d_zg[1], "gl": d_gl[0] + d_gl[1]}

#print(f"Manhattan distance from Zurich to Lugano is {manhattan["zl"]:.0f}")
#print(f"Manhattan distance from Zurich to Geneva is {manhattan["zg"]:.0f}")
#print(f"Manhattan distance from Geneva to Lugano is {manhattan["gl"]:.0f}")

# Euclidean distance

dx = cz[0] - cl[0]
dy = cz[1] - cl[1]

euclidean = {"zl": (d_zl[0]**2 + d_zl[1]**2) ** 0.5, "zg": (d_zg[0]**2 + d_zg[1]**2) ** 0.5, "gl": (d_gl[0]**2 + d_gl[1]**2) ** 0.5}

#print(f"Euclidean distance from Zurich to Lugano is {euclidean['zl']:.0f}")
#print(f"Euclidean distance from Zurich to Geneva is {euclidean['zg']:.0f}")
#print(f"Euclidean distance from Geneva to Lugano is {euclidean['gl']:.0f}")

# Distances Zurich to Lugano
distances = {
	"Zurich to Lugano": {"manhattan": manhattan["zl"], "euclidean": euclidean["zl"]},
	"Zurich to Geneva": {"manhattan": manhattan["zl"], "euclidean": euclidean["zl"]},
	"Genevaö to Lugano": {"manhattan": manhattan["zl"], "euclidean": euclidean["zl"]}
	}

print(distances)
```
