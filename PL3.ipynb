```python
# Part 1

# List of cities
cities = ["Zurich", "Geneva", "Lugano", "Regensberg"]

# List of coordinates
coordinates = [
	[2682217, 1247945],  # Zurich
	[2499959, 1117840],  # Geneva
	[2720031, 1098728],  # Lugano
	[2675257, 1259568],  # Regensberg
]

# Set up empty dictionary for above lists
city_data = {}

# Loop through cities and coordinates to create append to city_data
for i in range(len(cities)):
	city_data[cities[i]] = coordinates[i]

# Print coordinates of a city (here Geneva)
print(f"Geneva Coordinates: {city_data['Geneva']}")

# A dictionary is useful here, as it allows relating multiple variables with each other (i.e. name and coordinates), which can then be called together with a single call.
```

```python
# Part 2

# Loop through city_data to print each city with its coordinates
for city, coords in city_data.items():
	print(f"{city} lies at {coords[0]/1000:.0f} km east, {coords[1]/1000:.0f} km north")
```

```python
# Part 3

# Create list of city names
city_names = list(city_data.keys())

# Create empty list of city pairs
pairs = []

# Loop through city_data to create unique city pairs
for i in range(len(city_names)):
	for j in range(i + 1, len(city_names)):
		pairs.append((city_names[i], city_names[j]))

# Print all city pairs
print(pairs)

# Print number of unique city pairs
print(f"\nTotal unique pairs generated: {len(pairs)}")
```

```python
# Part 4

print(f"City pairs and their distances")

distances = {}

# Loop for calculating and listing city pairs and distances
for city_a, city_b in pairs:

	# Set up x and y variables
	xa, ya = city_data[city_a]
	xb, yb = city_data[city_b]

	# Calculate x and y distances
	dx = xa - xb
	dy = ya - yb
	
	# Calculate euclidean and manhattan distances
	euclidean = (dx**2 + dy**2) ** 0.5 / 1000
	manhattan = ((dx**2) ** 0.5 + (dy**2) ** 0.5)  / 1000

	# Create dictionary from distances
	distances[(city_a, city_b)] = {
		"euclidean": euclidean,
		"manhattan": manhattan
	}
	
	# Print city pairs and distances
	print(f"{city_a} — {city_b}: Euclidean = {euclidean:.0f} km, Manhattan = {manhattan:.0f} km")
```

```python
# Part 5

# Set up variables for the longest distance (euclidean and manhattan) and which city pair it is
max_euc = 0
max_euc_pair = None

max_man = 0
max_man_pair = None

# Iterate through city pair
for pair, metrics in distances.items():

	# Compare euclidean distances
	d_euc = metrics["euclidean"]
	if d_euc > max_euc:
		max_euc = d_euc
		max_euc_pair = pair
	
	# Compare manhattan distances
	d_man = metrics["manhattan"]
	if d_man > max_man:
		max_man = d_man
		max_man_pair = pair

# Print city pairs with the longest distance
print(f"Farthest cities (Euclidean): {max_euc_pair[0]} and {max_euc_pair[1]} with {max_euc:.0f} km")
print(f"Farthest cities (Manhattan): {max_man_pair[0]} and {max_man_pair[1]} with {max_man:.0f} km")
```

```python
# Part 6

# Set up a threshold of 200 km
threshold = 200

# Loop through pairs until distance exceeds 200 km
for city_a, city_b in pairs:
	
	# Set up x and y variables
	xa, ya = city_data[city_a]
	xb, yb = city_data[city_b]

	# Calculate x and y distances
	dx = xa - xb
	dy = ya - yb
	# Calculate euclidean distance
	distance = (dx**2 + dy**2) ** 0.5 / 1000
	if distance > threshold:
		print(f"Threshold exceeded by {city_a} and {city_b}: {distance:.0f}")
		break

# 1. Break will immediately stop the loop.
# 2. This is useful if you need at least one match. For example, if you need to find whether there is a hospital within a radius of a point, you can save calculation time/effort by stopping once you've found one.
```

```python
# Part 7

# Set up empty list for euclidean distances
euclidean_list = []

# Loop through distances to append euclidean distances to above list
for metrics in distances.values():
	euclidean_list.append(metrics["euclidean"])

# Sort euclidean list
euclidean_list.sort()

# Compute median
list_len = len(euclidean_list)

if list_len %2 == 0:
	j = list_len // 2
	median = (euclidean_list[j - 1] + euclidean_list[j]) / 2
else:
	j = list_len // 2
	median = euclidean_list[j]

# Print sorted distances and median
print(f"Sorted Distances: {[round(d, 0) for d in euclidean_list]}")
print(f"Median Distance: {median:.0f}")
```
